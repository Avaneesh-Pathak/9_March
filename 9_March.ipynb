Q1: The Probability Mass Function (PMF) and Probability Density Function (PDF) are mathematical functions used to describe the probability distribution of a random variable.

The PMF is used for discrete random variables and gives the probability of each possible outcome. It assigns probabilities to individual values. For example, consider a fair six-sided die. The PMF of this random variable would give the probability of rolling each of the numbers 1, 2, 3, 4, 5, and 6.

The PDF, on the other hand, is used for continuous random variables. It represents the likelihood of the random variable taking on a particular value within a range. Unlike the PMF, which assigns probabilities to specific values, the PDF assigns probabilities to intervals. For example, if we have a continuous random variable representing the height of individuals in a population, the PDF would describe the probability of a person having a height within a given range, such as between 160 cm and 170 cm.

Q2: The Cumulative Density Function (CDF) is a function that gives the probability that a random variable takes on a value less than or equal to a given value. It provides a cumulative view of the probabilities.

For example, let's say we have a random variable representing the time it takes for a customer to complete a transaction at a store. The CDF would give the probability that a customer takes less than or equal to a certain amount of time. If we want to find the probability that a customer takes less than 5 minutes to complete the transaction, we can use the CDF to calculate that probability.

The CDF is used to determine probabilities for a range of values and can be helpful in understanding the overall distribution of a random variable.

Q3: The normal distribution is commonly used as a model in various situations. Some examples include:

1. Heights of individuals in a population: The distribution of heights tends to follow a normal distribution with parameters that determine the mean and standard deviation. The mean represents the average height, and the standard deviation represents the spread or variability of heights in the population.

2. Errors in measurements: When taking measurements, there can be random errors involved. Assuming these errors are normally distributed can be a useful approximation. The mean of the normal distribution would represent the true value being measured, and the standard deviation would represent the uncertainty or variation in the measurements.

3. IQ scores: Intelligence quotient (IQ) scores are often assumed to follow a normal distribution. The mean represents the average IQ, and the standard deviation represents the spread of IQ scores around the mean.

In the normal distribution, the parameters determine the shape of the distribution. The mean (μ) determines the center or location of the distribution, while the standard deviation (σ) determines the spread or dispersion of the values. A larger standard deviation results in a wider distribution, while a smaller standard deviation leads to a narrower distribution.

Q5: The Bernoulli distribution is a discrete probability distribution that models a random experiment with two possible outcomes: success (usually represented by 1) and failure (usually represented by 0). It has a single parameter, usually denoted as p, which represents the probability of success.

An example of the Bernoulli distribution is modeling the outcome of a single coin toss, where success can be defined as getting heads (1) and failure as getting tails (0). The parameter p would represent the probability of getting heads.

The Binomial distribution, on the other hand, models the number of successes in a fixed number of independent Bernoulli trials. It is an extension of the Bernoulli distribution. The Binomial distribution has two parameters: n, the number of trials, and p, the probability of success in each trial.

For example, if we want to model the number of heads obtained in five coin tosses, we can use a Binomial distribution with n = 5 (number of trials) and p = 0.5 (probability of heads in each trial).

Q6: To calculate the probability that a randomly selected observation from a normally distributed dataset with a mean of 50 and a standard deviation of 10 is greater than 60, we need to use the Z-score and the standard normal distribution.

The Z-score is a measure of how many standard deviations an observation is away from the mean. It is calculated using the formula: Z = (X - μ) / σ, where X is the value of interest, μ is the mean, and σ is the standard deviation.

In this case, X = 60, μ = 50, and σ = 10. Plugging these values into the formula, we get: Z = (60 - 50) / 10 = 1.

Next, we can use a standard normal distribution table or a statistical software to find the probability associated with the Z-score of 1. From the standard normal distribution table, we find that the probability of a Z-score greater than 1 is approximately 0.1587.

Therefore, the probability that a randomly selected observation from the given dataset is greater than 60 is approximately 0.1587 or 15.87%.

Q7: The uniform distribution is a continuous probability distribution where all outcomes within a given interval have equal probabilities. It is characterized by a constant probability density function (PDF) over the interval.

For example, consider a random variable representing the time it takes for a student to complete a test. If we assume that the student is equally likely to take any amount of time between 0 and 60 minutes, we can model this using a uniform distribution. The PDF of the uniform distribution would be a constant value of 1/60 for the interval [0, 60], indicating that all values within this range have an equal probability of occurring.

Q8: The Z-score, also known as the standard score, is a measure used in statistics to standardize a value by expressing it in terms of standard deviations from the mean. It tells us how many standard deviations a particular observation is away from the mean and in which direction.

The formula for calculating the Z-score is: Z = (X - μ) / σ, where X is the value of interest, μ is the mean, and σ is the standard deviation.

The importance of the Z-score lies in its ability to provide a standardized value that can be compared across different distributions. By transforming values into Z-s

cores, we can compare and analyze data from different sources, regardless of their original units or scales. This allows us to make meaningful comparisons, identify outliers, and determine the relative position of an observation within a distribution.

Q9: The Central Limit Theorem (CLT) states that when independent random variables are added together, their sum tends to follow a normal distribution, regardless of the shape of the original variables' distributions. In other words, as the sample size increases, the sampling distribution of the mean approaches a normal distribution.

The significance of the Central Limit Theorem is that it provides a foundation for many statistical techniques and inference procedures. It allows us to make assumptions about the distribution of sample means, even when the population distribution is unknown or not normally distributed. It also provides a justification for using the normal distribution as an approximation in many real-world situations.

The Central Limit Theorem is particularly useful in hypothesis testing, confidence interval estimation, and constructing sampling distributions. It allows us to make reliable inferences about population parameters based on sample statistics.

Q10: The Central Limit Theorem relies on a few assumptions to hold true:

1. Independence: The random variables being combined or sampled are assumed to be independent of each other. This assumption ensures that the observations are not influenced by each other and that each observation provides new information.

2. Finite variance: The random variables have a finite variance, which measures the spread or variability of the variable. This assumption ensures that the sum or average of the variables remains well-behaved as the sample size increases.

3. Sample size: The Central Limit Theorem holds better as the sample size increases. While there is no fixed threshold, a general guideline is that the sample size should be reasonably large (often at least 30) for the theorem to provide a good approximation.

These assumptions, when met, allow the Central Limit Theorem to hold and enable the use of the normal distribution as an approximation for the sampling distribution of the mean. Violating these assumptions may lead to deviations from normality and affect the validity of the theorem's application.
